## Description of the Project :
This dataset is designed to study and diagnose **Erythemato-squamous diseases**, which include conditions like 

- Psoriasis  
- Seborrheic dermatitis  
- Lichen planus  
- Pityriasis rosea  
- Chronic dermatitis

For this study, patients were checked for 12 clinical features first. Then, skin samples were taken to study 22 features under a microscope.

**Attributes:**
The dataset has 34 attributes:

 - Clinical features (11 attributes): Observable symptoms like redness, scaling, itching, etc.
 - Histopathological features (22 attributes): Microscopic characteristics of skin samples, such as inflammation or thickened layers.
 - Other: Age of the patient.

**Scoring System**
- **Family history** is marked as `1` if the disease runs in the family, and `0` if not.  
- **Age** simply shows the patient’s age.  
- All other features are scored from `0` to `3`:  
  - `0`: symptom is not present.  
  - `1` or `2`: symptom is present at intermediate levels.  
  - `3`: symptom is present at its highest level.
 


## **Domain Analysis**

### **1. Problem Overview**

**Erythemato-squamous diseases (ESDs)** are a group of dermatological conditions that cause **inflammation** and **scaling** of the skin. These diseases commonly affect the **face, scalp**, and **extremities** (like the elbows and knees). They are difficult to diagnose because many of them share **similar clinical symptoms** such as redness (erythema), scaling, and itching. Additionally, their **histopathological features** (microscopic tissue changes) overlap significantly.

The six key diseases in this group are:

1. **Psoriasis**
2.  <img src="IMAGES/Psoriasis.webp" width="400" />
 - - What it is: A chronic skin condition where the skin cells build up too quickly, forming scales and itchy, dry patches.
 - - Common Symptoms: Red patches covered with silvery scales, dryness, cracking, and itching, often on the scalp, elbows, knees, or lower back.
 - - Cause: It’s an autoimmune condition, sometimes triggered by stress, infections, or cold weather.



3. **Seborrheic Dermatitis**
4. <img src="IMAGES/Seorrheic dermatitis.jpg" width="400" />
 - - What it is: A common skin condition causing redness, scaly patches, and dandruff.
 - - Common Symptoms: Itchy, flaky skin on the scalp, face (around the nose and eyebrows), or chest.
 - - Cause: Likely caused by a combination of genetics, stress, and yeast growth on the skin.



5. **Lichen Planus**
6. <img src="IMAGES/Lichen-Planus.jpg" width="400" />
 - - What it is: An inflammatory skin condition that can affect the skin, mouth, or nails.
 - - Common Symptoms: Flat, purple, itchy bumps on the skin, often on the wrists, legs, or lower back. In the mouth, it can cause white, lacy patches or sores.
 - - Cause: Often linked to an overactive immune system; the exact cause is unclear.



4. **Pityriasis Rosea**
5. <img src="IMAGES/Pityriais_rosea.jpg" width="400" />
- - What it is: A temporary skin rash that typically starts with one large patch and spreads smaller patches.
 - - Common Symptoms: A "herald patch" (large, single spot) followed by smaller, oval-shaped spots on the chest, back, or abdomen.
 - - Cause: Likely caused by a viral infection but not contagious.



8. **Chronic Dermatitis**
9. <img src="IMAGES/eczema_original.webp" width="400" />
 - - What it is: Long-term inflammation of the skin that causes redness, swelling, and itching.
 - - Common Symptoms: Thickened, scaly skin due to repeated scratching, often on the hands, feet, or areas prone to irritation.
 - - Cause: Triggered by allergens, irritants, or underlying conditions like eczema or allergies.



10. **Pityriasis Rubra Pilaris**
11. <img src="IMAGES/pityriasis rubra pilaris.jpg" width="400" />
 - - What it is: A rare skin disorder causing inflammation and scaling of the skin.
 - - Common Symptoms: Orange-red patches with thickened skin and small, hard bumps, often on the palms, soles, and scalp.
 - - Cause: Unknown in most cases, but it may be genetic or linked to an immune system issue.



The challenge in dermatology is accurately diagnosing these diseases since they look similar, especially in the early stages. Traditional methods involve visual examination, biopsy, and a detailed review of symptoms. However, the overlapping features often make a clear distinction difficult. Thus, there is a need for an **automated system** that can analyze clinical and histopathological features to assist in diagnosing these diseases efficiently.

---

### **4. Challenges in the Domain**

1. **Similar Symptoms Across Diseases**:
   - ESDs share common clinical features, such as erythema, scaling, and itching, making it hard to distinguish between them. Even biopsy samples may not clearly differentiate them, as many share similar histopathological features.

2. **Biopsy Complexity**:
   - While biopsy provides critical information, the **microscopic features** are often subtle, and the diseases share many common histopathological traits, which complicates diagnosis.

3. **Class Imbalance**:
   - Some diseases (e.g., Psoriasis) are more common than others (e.g., Pityriasis Rubra Pilaris), leading to **imbalanced classes**. This could cause the model to be biased toward more common diseases.

4. **Data Quality**:
   - **Missing data** or **inconsistent feature values** may occur in the dataset, especially for clinical observations. This could affect the model’s performance and require careful data preprocessing.

5. **Overlapping Disease Features**:
   - At early stages, one disease may exhibit features typical of another, which makes it even harder for clinicians to differentiate based on symptoms alone.

---

### **5. Solution Approach with Machine Learning**

The goal is to develop a **classification model** that accurately predicts the disease based on the provided clinical and histopathological features. The approach will involve:
- **Data preprocessing**: Handling missing values, encoding categorical variables, and normalizing features where necessary.
- **Model training**: Using algorithms like decision trees, random forests, or neural networks to train the model on the input features.
- **Model evaluation**: Evaluating the model’s accuracy, precision, recall, and F1-score using a testing set or cross-validation techniques to ensure the model generalizes well.
- **Deployment**: Once validated, the model can assist dermatologists by providing an automated tool for diagnosis, reducing diagnostic time and improving accuracy.

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, recall_score, confusion_matrix, classification_report
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],  # Inverse of regularization strength
    'solver': ['lbfgs', 'liblinear'],  # Solvers for optimization
    'penalty': ['l2']  # Regularization penalty
}

grid_search = GridSearchCV(estimator=LogisticRegression(max_iter=1000), param_grid=param_grid, cv=3, scoring='accuracy')
grid_search.fit(X_train, y_train)

print(f"Best Parameters: {grid_search.best_params_}")
```

```python
lr = LogisticRegression(multi_class="ovr",C=1,solver="lbfgs")
lr.fit(X_train,y_train)
```

```python
y_pred_train_LR = lr.predict(X_train)
y_pred_test_LR = lr.predict(X_test)

# Function to calculate and print metrics
def print_metrics(y_true, y_pred, dataset_type="Dataset"):
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted')
    recall = recall_score(y_true, y_pred, average='weighted')

    print(f"{dataset_type} Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"Recall: {recall:.4f}")
  
    # Confusion Matrix
    conf_matrix = confusion_matrix(y_true, y_pred)
    print(f"Confusion Matrix ({dataset_type}):\n{conf_matrix}\n")
    
    # Classification Report
    print(classification_report(y_true, y_pred))
    print("-----------------------------------------------------")


# Print metrics for training and testing datasets
print_metrics(y_train, y_pred_train_LR, "Training")
print_metrics(y_test, y_pred_test_LR, "Testing")

# Comparison at the end
print("Training vs Testing Performance:")
print(f"Training Accuracy: {accuracy_score(y_train, y_pred_train_LR):.4f} | Testing Accuracy: {accuracy_score(y_test, y_pred_test_LR):.4f}")
print(f"Training F1 Score: {f1_score(y_train, y_pred_train_LR, average='weighted'):.4f} | Testing F1 Score: {f1_score(y_test, y_pred_test_LR, average='weighted'):.4f}")
```
